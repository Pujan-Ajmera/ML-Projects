**The process where we convert catogerical data into numerical one is called the ENCODING:

1)One Hot Encoding :used when we have data with just 2-3 sub categories like male/ female or 
yes/no or rural/semi-rural/urban  this type of data.

# say we have male and female so what happens is that we first get to make 2 colums 
1) gender_male
2) gender_female
 then this 2 colums are filled using 0 or 1 and its done in a way such that if the gender column has male the gender_male gets 1 and gender_female gets 0 and so on for the whole column 
 

```python
import pandas as pd
```

```python
dataset =  pd.read_csv(r"D:/PLAYGROUND/ml/datasets/loan.csv")
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset["Married"]=dataset["Married"].fillna(dataset["Married"].mode()[0])
```

# there are 2 ways to perform one hot encoding:
1) GetDummies methods using pandas
2) OneHotEncodder class in sclearn

```python
en_data = dataset[["Gender","Married"]]
```

```python
pd.get_dummies(en_data)
```

# now this data is in the form of T/F and this data needs to be converted to numbers but the other way is to do by using sclearn:

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
# what it does is that it see the data first then it applies its algo to encode data
# like conversion into a numerical data from a catogerial datasets
```

```python
ar = ohe.fit_transform(en_data).toarray()
# sparse matrix most elements are 0 as in DSA
# ohe.fit_transform(en_data)  this gives a sparsh matrix 
print(ar)
```

```python
# cnoverting this data to a DataFrame
pd.DataFrame(ar, columns=ohe.get_feature_names_out(["Gender", "Married"]))
```

# Ordinal Encoding: where there is some relation in colums
1) By scikit learn
2) by map function

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df
```

```python
ord_data = [["s","m","l","xl"]]
ord_data
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"]=oe.transform(df[["Size"]])
df["Size_en"]
df
```

```python
# using map function
```

```python
ord_data1 = {"s":5,"m":6,"l":7,"xl":8}
```

```python
df["Size_en_map"] = df["Size"].map(ord_data1)
df
```
