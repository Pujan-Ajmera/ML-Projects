# three types of data filling 
# 1)MOD filling - here the most frequent type of data are filled
# 2)BACKWARD filling - here the behind data are filled in front
# 3)FORWARD filling - this is the filling where we fill data from forward to backward like just the previous value of a null value is filled in this
# null value if that value is not null else as it is same for backward filling

```python
# dataset.fillna(method="bfill",axis=0)
# this methods are deplicated like method param = "f/bfill" so use the below things
dataset.ffill(axis=0) 
# here the axis is the row or column and 0 =  row and 1 = column
# the meaning is that the 0 when used uses the rows previous or the forward data to fill missing data 
# the meaning is that the 1 when used uses the columns previous or the forward data to fill missing data 
```

```python
# MOD Data filling is only done for a single column
# dataset["Gender"].mode()[0] this gives the mode value
dataset["Gender"] = dataset["Gender"].fillna(dataset["Gender"].mode()[0])
# here the data is changed inplace as the inplace thing is not working as its maybe depricated in this caseabs
```

```python
for i in dataset.select_dtypes(include="object").columns:
    print(i)
    dataset[i] = dataset[i].fillna(dataset[i].mode()[0])
```
