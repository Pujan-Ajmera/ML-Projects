```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv(r"D:/PLAYGROUND/ml/datasets/loan.csv")
dataset.head(3)
```

```python
dataset['ApplicantIncome']=dataset['ApplicantIncome'].fillna(dataset['ApplicantIncome'].mean())
dataset['ApplicantIncome']
```

```python
sns.displot(dataset['ApplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome'] = pd.DataFrame(ss.transform(dataset[['ApplicantIncome']]),columns=['x'])
```
